### 데이터셋 생성

```python
import numpy as np
import pandas as pd

student = ['야스오', '오공', '블리츠', '페이커']

math_study_time = [2, 4, 6, 8]
eng_study_time = [3, 8, 2, 4]
sci_study_time = [1, 9, 7, 6]
score = [3, 7, 5, 8]

dataset = pd.DataFrame({'math':math_study_time, 'english':eng_study_time, 'science':sci_study_time, 'score':score}, index=student)
dataset.head()
```

```python
y_train = dataset['score'].to_numpy()
X_train = dataset.drop('score', axis=1).to_numpy()
print(X_train, y_train)
```

### 가설함수 설정
w1x1 + w2x2 + w3x3 + b

```python
W = np.ones((3, 1))
print(W)
```

```python
from random import *

# b = np.random.rand(1, 1)
b = np.ones((1, 1))
print(b)
```

```python
def f(X):
    return np.dot(X, W) + b
```

```python
y_pred = f(X_train)
print(y_pred)
```

### Cost Function

```python
def loss_function(y, y_pred):
    n_samples = len(y_pred)
    deviation = y_pred - y
    cost = np.sum(np.square(deviation)) / n_samples
    return cost
```

```python
loss_function(y_train, y_pred)
```

### Weight, Bias Update

```python
def gradient(X_train, y_train, y_pred):
    n_samples = len(y_train)
    dw = (1/n_samples) * np.dot(X_train.T, (y_pred - y_train))
    return dw
```

```python
dw = gradient(X_train, y_train, y_pred)
```

```python
def gradient_descent(W, learning_rate, dw):
     W = W - learning_rate * dw
     return W
```

```python
W = gradient_descent(W, 0.01, dw)
print(W)
```

```python
losses = []

for i in range(100):
    y_pred = f(X_train)

    loss = loss_function(y_train, y_pred)
    losses.append(loss)

    dw = gradient(X_train, y_train, y_pred)
    W = gradient_descent(W, 0.01, dw)

    print('W: {}, loss: {}'.format(W, loss))
```
